### Load Data (3 Marks)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import tensorflow as tf
import os
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import  Embedding, LSTM, Bidirectional, BatchNormalization, Dense, Dropout, GlobalMaxPool1D
from tensorflow.keras import Sequential
```

```python
from google.colab import drive 
drive.mount('/content/drive')
path = os.listdir('/content/drive/My Drive/NLP/sarcasm/')

data_json = pd.read_json('/content/drive/My Drive/NLP/sarcasm/Sarcasm_Headlines_Dataset.json',lines = True, orient = 'values')
print("Json file shape: " , data_json.shape)
data_json[:5]
```

### Drop `article_link` from dataset (3 Marks)

```python
data = data_json.drop(labels = 'article_link', axis = 1)
```

### Get length of each headline and add a column for that (3 Marks)

```python
data['headline_length'] = [len(x.split(" ")) for x in data['headline']]
```

```python
data.head()
```

### Initialize parameter values
- Set values for max_features, maxlen, & embedding_size
- max_features: Number of words to take from tokenizer(most frequent words)
- maxlen: Maximum length of each sentence to be limited to 25
- embedding_size: size of embedding vector

```python
max_features = 10000
maxlen = 25
embedding_size = 200
```

### Apply `tensorflow.keras` Tokenizer and get indices for words (3 Marks)
- Initialize Tokenizer object with number of words as 10000
- Fit the tokenizer object on headline column
- Convert the text to sequence


```python
texts = list(data['headline'])
tokenizer = Tokenizer(num_words = max_features, filters='!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n', lower=True,split = " ")
tokenizer.fit_on_texts(texts)
text_to_sequence = tokenizer.texts_to_sequences(texts)
```

### Pad sequences (3 Marks)
- Pad each example with a maximum length
- Convert target column into numpy array

```python
padded_text = pad_sequences(text_to_sequence, maxlen=maxlen)
X = padded_text
y =np.asarray(data['is_sarcastic'])
```

### Vocab mapping
- There is no word for 0th index

```python
tokenizer.word_index
```

### Set number of words
- Since the above 0th index doesn't have a word, add 1 to the length of the vocabulary

```python
num_words = len(tokenizer.word_index) + 1
print(num_words)
```

### Load Glove Word Embeddings (3 Marks)

```python
open('/content/drive/My Drive/NLP/sarcasm/glove.6B.200d.txt', 'r')
```

### Create embedding matrix

```python
EMBEDDING_FILE = '/content/drive/My Drive/NLP/sarcasm/glove.6B.200d.txt'

embeddings = {}
for o in open(EMBEDDING_FILE):
    word = o.split(" ")[0]
    # print(word)
    embd = o.split(" ")[1:]
    embd = np.asarray(embd, dtype='float32')
    # print(embd)
    embeddings[word] = embd

# create a weight matrix for words in training docs
embedding_matrix = np.zeros((num_words, 200))

for word, i in tokenizer.word_index.items():
	embedding_vector = embeddings.get(word)
	if embedding_vector is not None:
		embedding_matrix[i] = embedding_vector
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, shuffle = True)
```

### Define model (5 Marks)
- Hint: Use Sequential model instance and then add Embedding layer, Bidirectional(LSTM) layer, then dense and dropout layers as required. 
In the end add a final dense layer with sigmoid activation for binary classification.

```python
model = Sequential()
model.add(Embedding(input_dim = len(tokenizer.word_index)+1, output_dim = embedding_size, input_length = maxlen, weights =[embedding_matrix], trainable = False))
model.add(Bidirectional(LSTM(128, return_sequences= True)))
model.add(Dense(256, activation = 'relu'))
model.add(Dropout(0.3))
model.add(Dense(1, activation = 'sigmoid'))
model.summary()
```

### Compile the model (3 Marks)

```python
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

### Fit the model (4 Marks)

```python
model.fit(X_train,y_train, batch_size = 512, epochs = 10, validation_split = 0.1)
```

```python
model.evaluate(X_test, y_test)
```
