```python
import numpy as np
import pandas as pd
```

```python
#Reading the dataset
la_liga = pd.read_csv("Laliga.csv", header = 1 )
```

```python
#Getting basic information from the dataset
```

```python
#Checking null values
la_liga.isnull().sum()
```

```python
#Replacing '-' with 0
la_liga.replace('-',0, inplace=True)
```

```python
#checking datatype
la_liga.dtypes
```

```python
#Copying the info from 'Debut' column to 'Debut mod' column and slicing it to remove the data 
#after '-' and converting it to int 
```

```python
#la_liga['Debut']=la_liga['Debut'].astype(str)
la_liga['Debut mod'] = la_liga['Debut'].str.slice(0,4)
la_liga['Debut mod'] = la_liga['Debut mod'].astype(int)
```

```python
#Info of teams who played between 1930 and 1980. Did not include 1929-30 since that would mean the teams 
#started in the year 1929
la_liga_debut = la_liga.loc[(la_liga['Debut mod']>=1930)&(la_liga['Debut mod']<=1980)]

print(la_liga_debut.iloc[:,1].to_string(index=False))
```

```python
#Converting datatype of 'Points' to int from object
la_liga['Points'] = la_liga['Points'].astype(int)
```

```python
#Sorting based on points
la_liga.sort_values(by=['Points'], ascending=False, inplace=True)
```

```python
#Printing a list of Top 5 teams based on points
print(la_liga.iloc[:5,1].values.tolist())
```

```python
#Goal_diff_count returns a dataframe with team and their goal difference and also prints the maximum goal 
#difference and the team

def Goal_diff_count():
    goal_diff = la_liga['GoalsFor'].astype(int) - la_liga['GoalsAgainst'].astype(int)
    goal_diff_dict = {'Team': la_liga['Team'],'Goal_diff': goal_diff}
    Goal_diff_df = pd.DataFrame(goal_diff_dict)
    print(f"Max goal difference :{Goal_diff_df.loc[Goal_diff_df['Goal_diff'].idxmax()].to_string(index=False)}")
    print(f"Min goal difference :{Goal_diff_df.loc[Goal_diff_df['Goal_diff'].idxmin()].to_string(index=False)}")
    return Goal_diff_df

Goal_diff_count()
```

```python
#Creating new column 'Winning percentage'
la_liga['Winning Percent'] = (la_liga['GamesWon'].astype(int)/la_liga['GamesPlayed'].astype(int))*100
```

```python
#Replacing NaN with 0
la_liga['Winning Percent'].fillna(0,inplace=True)
```

```python
#Printing top 5 teams by winning percent
la_liga.sort_values(by=['Winning Percent'], ascending=False, inplace=True)
print(la_liga.iloc[:5,1].to_string(index=False))
```

```python
#Grouping by best position and summing the points
la_liga.groupby(['BestPosition']).sum()[['Points']]
```
